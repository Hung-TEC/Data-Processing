```python
def plot_b(name):
    if(name == 'calendar.csv'):
        df=cal_df
        for i in df.columns:
            df_vc = df[i].value_counts()
            number_of_col = (df_vc/len(df)).sort_index()
            number_of_col.plot(kind='bar',figsize=(9,6))
            L = len(number_of_col)
            if L > 100:
                plt.xticks(range(0, L, 50),
                           number_of_col.index.tolist()[0:L:50])
            plt.legend([i])
            plt.xlabel(i)
            plt.ylabel("Frequency")
            plt.title("Frequency graphic of "+i)
            plt.show()
    elif(name == 'reviews.csv'):
        df=rview_df
        for i in ['listing_id', 'date']:
            df_vc = df[i].value_counts()
            number_of_col = (df_vc/len(df)).sort_index()
            number_of_col.plot(kind='bar',figsize=(9,6))
            L = len(number_of_col)
            if L > 100:
                plt.xticks(range(0, L, 100),
                           number_of_col.index.tolist()[0:L:100])
            plt.legend([i])
            plt.xlabel(i)
            plt.ylabel("Frequency")
            plt.title("Frequency graphic of "+i)
            plt.show()
```

### 3. Task b


We are using KNeighbors Regression as the machine learning model to forecast the price according to user requirement.

In Supervised Learning, KNeighbors Regression is quite best for working with continious and discrete value. KNeighbors regression is a technique used when the data is faced with independent correlation variables. 

KNeighbors regression could Use the target values of the K most recent training samples to make a decision on the regression values of the test samples. In addition, KNeighbors Regression limits the radius of the nearest neighbor leaf of tree and the class of the nearest neighbor regression tree with a limited radius, as well as the determination of the nearest centioria, so that this approach can perfectly reduce the error brought by the large variance to the model itself.

Meanwhile, KNeighbors regression could perform many task like 

1.   Forcasting(Price or any other variable)
2.   Forcasting Trend

Can you tune your model to perform better?

Explain the technique you’re using and justify why it is improving your results. 
    
    Here we use the grid search model to help us adjust the most suitable parameters for each original parameter in the knn regression model. Our method is to first give some preset parameters and detect the most suitable for these parameters. Model, the best training score to help us determine which is the best model, and output their best original parameters.

```python
from sklearn.model_selection import GridSearchCV


def build_model(datas, target):
    x_train, x_test, y_train, y_test = train_test_split(
        datas, target, train_size=0.9,random_state=5)
    knn = KNeighborsRegressor()
#     model=knn
    parameters_grid= {'C':[0.0001,0.01,0.1,1,3,6,5,20],
          'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
          'leaf_size':[30,50,100,200,5,3],
          'p':[1,2],
           'metric':['minkowski','str', 'callable']      
                     }
    model=GridSearchCV(knn,param_grid=parameters_grid,cv=5)
    model.fit(x_train,y_train)
    print('model.best_params_::  ',model.best_params_)
    pre=model.predict(x_train)
    print('Best score for the model：',r2_score(y_train,pre))
#     y_train=np.array(y_train).reshape(1,-1)

build_model(datas, target)

```
